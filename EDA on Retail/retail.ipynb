```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("SampleSuperstore.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
print(df.isnull().sum())
```

```python
df = df.drop(['Postal Code'], axis=1)
```

```python
print("Duplicates:", df.duplicated().sum())
```

```python
df = df.drop_duplicates()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Category', y='Sales', data=df)
plt.title('Sales by Category')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Category', y='Profit', data=df)
plt.title('Profit by Category')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='Region', y='Sales', data=df)
plt.title('Sales by Region')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='Profit', data=df)
plt.title('Profit by Region')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='Sub-Category', y='Sales', data=df)
plt.title('Sales by Sub-Category')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(15, 9))
sns.barplot(x='Sub-Category', y='Profit', data=df)
plt.title('Profit by Sub-Category')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.lineplot(x='Discount', y='Profit', data=df)
plt.title('Impact of Discounts on Profit')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Category', data=df)
plt.title('Returns by Category')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Quantity', y='Profit', data=df, hue='Category')
plt.title('Profit vs Quantity Sold')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Ship Mode', data=df, hue='Segment')
plt.title('Shipping Mode Analysis by Segment')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['Discount'], bins=20, kde=True)
plt.title('Distribution of Discounts')
plt.xlabel('Discount')
plt.show()
```
